```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Define the file path to the Excel file in Google Drive
file_path = "/content/drive/MyDrive/Ciencia de datos/corte 2/reto/datos/Tabela 5457.xlsx"

# Load the Excel file into a pandas DataFrame
try:
    df_original = pd.read_excel(file_path)
    print("DataFrame original cargado exitosamente:")
    display(df_original.head())
except FileNotFoundError:
    print(f"Error: The file was not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")

# Extract years and their corresponding starting column indices in the original DataFrame (based on row 2)
years_with_indices_original = [(year, i) for i, year in enumerate(df_original.iloc[2].tolist()) if pd.notna(year)]
print("\nAños con sus índices de columna iniciales en el DataFrame original (basado en la fila 2):")
print(years_with_indices_original)

# Extract product names from row 3 (excluding the first two columns which are not products)
product_names_row_3 = df_original.iloc[3, 2:].tolist()
print("\nNombres de productos extraídos de la fila 3:")
print(product_names_row_3)

# The 'Geographical Region' data is in the first column, starting from row 4.
geo_region_data = df_original.iloc[4:, 0].reset_index(drop=True)

melted_data = []

# Iterate through each year and its corresponding starting column index
for i, (year, year_col_index) in enumerate(years_with_indices_original):
    # Determine the range of columns for the current year's data in the original DataFrame.
    # The start of the current year's block is the column index from row 2.
    # The end of the current year's block is the start of the next year's block, or the end of the DataFrame if it's the last year.
    if i < len(years_with_indices_original) - 1:
        next_year_col_index = years_with_indices_original[i+1][1]
    else:
        next_year_col_index = df_original.shape[1] # End of the DataFrame

    # Within this block of columns for the current year, we need to identify the column
    # that contains the 'Cantidad producida (Toneladas)' for each product.
    # Based on the structure, it seems the columns within each year block are ordered
    # consistently for each product, and the column with the product name in row 3
    # corresponds to the 'Cantidad producida (Toneladas)'.

    # Find the column indices within the original DataFrame for the current year's data
    # that correspond to the product names from row 3 and contain the production quantity.
    current_year_product_quantity_cols = []
    # Iterate through the columns within the current year's block
    for col_index in range(year_col_index, next_year_col_index):
        product_in_header_row_3 = df_original.iloc[3, col_index]
        # Check if the header in row 3 is a product name (i.e., not NaN, and not 'Total' or other non-product headers)
        if pd.notna(product_in_header_row_3) and product_in_header_row_3 in product_names_row_3:
             # This column corresponds to a product within the current year's block and contains the production quantity.
             current_year_product_quantity_cols.append({
                 'Product': product_in_header_row_3,
                 'Original_Column_Index': col_index
             })

    # Now, extract the data for the identified production quantity columns for the current year.
    year_melted_data = []
    for col_info in current_year_product_quantity_cols:
        product = col_info['Product']
        original_col_index = col_info['Original_Column_Index']

        # Extract the data for this column, starting from row 4
        product_data = df_original.iloc[4:, original_col_index].reset_index(drop=True)

        # Create a temporary DataFrame for this product and year
        temp_df = pd.DataFrame({
            'Geographical Region': geo_region_data,
            'Year': year,
            'Product': product,
            'Value': product_data
        })

        year_melted_data.append(temp_df)

    # Concatenate the melted data for the current year
    if year_melted_data: # Check if there's any data for the current year
        year_melted_df = pd.concat(year_melted_data, ignore_index=True)
        melted_data.append(year_melted_df)


# Concatenate all the melted dataframes for all years
df_melted = pd.concat(melted_data, ignore_index=True)


# Convert 'Value' column to numeric, coercing errors to pd.NA
df_melted['Value'] = pd.to_numeric(df_melted['Value'], errors='coerce')

# Replace placeholder values in the 'Value' column after conversion
df_melted['Value'] = df_melted['Value'].replace(['..', '-'], pd.NA)


# Drop rows where 'Value' is NA (as these represent missing production data)
df_melted.dropna(subset=['Value'], inplace=True)

# Sort by year
df_sorted_by_year = df_melted.sort_values(by='Year')

print("\nDataFrame final limpio y reestructurado:")
display(df_sorted_by_year.head())
print("\nInformación del DataFrame final:")
df_sorted_by_year.info()


# --- Data Analysis and Visualization ---

print("\n## Análisis Estadístico de la Base de Datos\n")

# 1. Inspeccionar Datos
print("### Inspección de Datos\n")
print("Primeras filas del DataFrame:")
display(df_sorted_by_year.head())
print("\nInformación del DataFrame:")
df_sorted_by_year.info()
print("\nEstadísticas Descriptivas:")
display(df_sorted_by_year.describe())

# 2. Identificar Productos Principales
print("\n### Productos Principales\n")
total_production_by_product = df_sorted_by_year.groupby('Product')['Value'].sum().sort_values(ascending=False)
top_10_products = total_production_by_product.head(10)
print("Los 10 Productos Principales por Producción Total:")
display(top_10_products)

# 3. Analizar Tendencias de Producción a lo Largo del Tiempo
print("\n### Tendencias de Producción a lo Largo del Tiempo\n")
total_production_by_year = df_sorted_by_year.groupby('Year')['Value'].sum()
print("Producción Total por Año:")
display(total_production_by_year.head())

# Filter for top 10 products and calculate total production by year and product
df_top_10_products = df_sorted_by_year[df_sorted_by_year['Product'].isin(top_10_products.index)]
total_production_top_10_by_year_product = df_top_10_products.groupby(['Year', 'Product'])['Value'].sum().reset_index()
print("\nProducción Total de los 10 Productos Principales a lo Largo de los Años:")
display(total_production_top_10_by_year_product.head())


# 4. Analizar Producción por Región Geográfica
print("\n### Producción por Región Geográfica\n")
total_production_by_region = df_sorted_by_year.groupby('Geographical Region')['Value'].sum().sort_values(ascending=False)
print("Producción Total por Región Geográfica:")
display(total_production_by_region.head())

# 5. Manejar Valores Faltantes
print("\n### Manejo de Valores Faltantes\n")
missing_values_count = df_sorted_by_year['Value'].isna().sum()
total_values_count = len(df_sorted_by_year['Value'])
percentage_missing = (missing_values_count / total_values_count) * 100

print(f"Número de valores faltantes en la columna 'Value': {missing_values_count}")
print(f"Número total de valores en la columna 'Value': {total_values_count}")
print(f"Porcentaje de valores faltantes en la columna 'Value': {percentage_missing:.2f}%")

# 6. Visualizar Hallazgos Clave
print("\n### Visualización de Hallazgos Clave\n")

# 1. Gráfico de Líneas para la Producción Total a lo Largo del Tiempo
plt.figure(figsize=(12, 6))
total_production_by_year.plot(kind='line', marker='o')
plt.title("Producción Agrícola Total a lo Largo del Tiempo")
plt.xlabel("Año")
plt.ylabel("Producción Total")
plt.grid(True)
plt.show()

# 2. Gráfico de Barras Apiladas para la Producción de los 10 Productos Principales a lo Largo del Tiempo
plt.figure(figsize=(15, 8))
# Pivot the data for stacked bar chart
pivot_df = total_production_top_10_by_year_product.pivot(index='Year', columns='Product', values='Value')
pivot_df.plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title("Producción de los 10 Productos Principales a lo Largo del Tiempo")
plt.xlabel("Año")
plt.ylabel("Producción Total")
plt.legend(title="Producto", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 3. Gráfico de Barras para la Producción Total por Región Geográfica
plt.figure(figsize=(12, 6))
total_production_by_region.plot(kind='bar')
plt.title("Producción Agrícola Total por Región Geográfica")
plt.xlabel("Región Geográfica")
plt.ylabel("Producción Total")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 7. Resumir Hallazgos (ya incluido en los print statements)
print("\n## Resumen del Análisis de Datos\n")

print("### Hallazgos Clave:\n")
print(f"- **Producción Total:** '{top_10_products.index[0]}' es el producto principal por producción total en todos los años y regiones.")
print("- **Tendencias Temporales:** Se observan tendencias de producción variables a lo largo del tiempo para los diferentes productos.")
print(f"- **Distribución Regional:** '{total_production_by_region.index[0]}' es la región geográfica con la mayor producción agrícola total.")
print(f"- **Valores Faltantes:** Aproximadamente el {percentage_missing:.2f}% de la columna 'Value' contiene datos faltantes.")

print("\n### Visualizaciones Destacadas:\n")
print("- Se mostró un gráfico de líneas para la producción agrícola total a lo largo del tiempo.")
print("- Un gráfico de barras apiladas ilustró la producción de los 10 productos principales a lo largo de los años.")
print("- Un gráfico de barras mostró la producción agrícola total por región geográfica.")

print("\n### Ideas y Próximos Pasos:\n")
print("Los datos han sido limpiados, organizados y analizados para identificar productos principales, tendencias temporales y distribución regional. Se han visualizado los hallazgos clave. Próximos pasos podrían incluir un análisis más profundo de las causas de los valores faltantes o la incorporación de otros conjuntos de datos.")
```
